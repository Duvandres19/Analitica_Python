```python
import csv
 
with open('../fuentes/lecturaCSV.csv') as products:
    leer = csv.reader(products, delimiter=';', quotechar=',',
                        quoting=csv.QUOTE_MINIMAL)
    for row in leer:
        print("ID: {0},Codigo orden: {1}, Fecha Orden: {2}".format(row[0],row[1],row[2]))
```
